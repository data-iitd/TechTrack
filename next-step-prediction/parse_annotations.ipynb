```python
import pickle
from collections import namedtuple
 
Transition = namedtuple('Transition',['entity', 'property', 'value'])
```

```python
with open('all_files_v1.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
property_event = {}
property_event['notRelevant'] = True
property_event['isPowered'] = False
property_event['isConnected'] = False
property_event['isSetup'] = True
property_event['isUsed'] = True
property_event['isInstalled'] = False
property_event['isRelatedDeviceConnected'] = False
property_event['isSettingsChanged'] = True
property_event['isOpened'] = False

property_values = {}
property_values['True'] = True
property_values['False'] = False
```

```python
_entities, _properties, _values = {}, {}, {}
_transitions = {}
_dropped_count, _total_count = 0, 0

DATASET = {}

for procedure, annotations in data.items():
    # print(procedure)
    steps = [[] for i in range(len(annotations['steps']))]
    
    for entity, prop_val in annotations['step_wise_properties'].items():
        for property, values in prop_val.items():
            # print(entity, property, values, property_event[property])
            if property=='notRelevant':
                continue
            if property_event[property]:
                for index, value in enumerate(values):
                    if property_values[value] == True:
                        transition = Transition(entity=entity,property=property,value=value)
                        # print(index, transition)
                        steps[index].append(transition)
                        
                        _entities[entity] = _entities.get(entity, 0) + 1
                        _properties[property] = _properties.get(property, 0) + 1
                        _values[value] = _values.get(value, 0) + 1
                        _transitions[transition] = _transitions.get(transition, 0) + 1
            else:
                prev_val = False
                for index, value in enumerate(values):
                    if property_values[value] != prev_val:  # transition
                        transition = Transition(entity=entity,property=property,value=value)
                        # print(index, transition)
                        steps[index].append(transition)
                        
                        _entities[entity] = _entities.get(entity, 0) + 1
                        _properties[property] = _properties.get(property, 0) + 1
                        _values[value] = _values.get(value, 0) + 1
                        _transitions[transition] = _transitions.get(transition, 0) + 1
                        
                        prev_val = property_values[value]
                
    processed_steps = []
    drop = False
    for step in steps:
        if len(step) == 0:
            continue
        #elif len(step) == 1:
        else:
            processed_steps.append(step)
        #else:
        #    drop=True
        #    break

    if len(processed_steps) == 0:
        print(procedure)
        # print(steps)
        _dropped_count += 1
    elif not drop:
        DATASET.update({procedure: processed_steps.copy()})
    else:
        print(procedure)
        # print(steps)
        _dropped_count += 1
    
    _total_count += 1
        
        
```
